```python
import keras.utils as image
import numpy as np

img = image.load_img("./Lenna.png").convert("RGB")
img = np.asarray(img)/255

greyscale = np.asarray([ 
            [[np.mean(pixel)] for pixel in col]
             for col in img
            ])
            #  for pixel in col 
image.array_to_img(greyscale).save('Lenna-grey.png')

threshold = np.asarray([col.sum() for col in greyscale]).sum() / (512*512)
binary = np.asarray([
        [[int(pixel > threshold)] for pixel in col] 
        for col in greyscale
        ])
image.array_to_img(binary).save('Lenna-binary.png')
```
